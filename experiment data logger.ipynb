```python
# Initialize a list to hold your experiment data if you haven't already
if 'experiment_logs' not in globals():
    experiment_logs = []

# Define your new data as a list of dictionaries
new_trials = [
    {"experiment_name": "Trial 3 - Light Intensity", "parameter": "Lux", "value": 450, "notes": "Indoor lighting"},
    {"experiment_name": "Trial 4 - Battery Drain", "parameter": "Current (mA)", "value": 120.5, "notes": "Active mode"}
]

# Process and log them
for entry in new_trials:
    log_data(**entry)
    experiment_logs.append(entry)

print(f"Successfully logged {len(new_trials)} new trials.")
```

```python
import pandas as pd

# Assuming 'experiment_logs' is the list where you've been storing your data
# If you are pulling this from a database or a specific log file, 
# ensure you have that data stored in a variable named 'experiment_logs'

# Convert the log data into a pandas DataFrame
df = pd.DataFrame(experiment_logs)

# Save the DataFrame to a CSV file
file_name = "experimental_data_log.csv"
df.to_csv(file_name, index=False)

print(f"Data successfully exported to {file_name}")
```
